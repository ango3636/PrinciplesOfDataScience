```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
# read in clean data
clean_student_data = pd.read_csv("/Users/amyngo/Desktop/5530 Data Science/A1-AmyNgo/Student-Performance-Visuals/data-clean/clean_student_data.csv")
clean_student_data.head(5)

#save directory
save_folder = '/Users/amyngo/Desktop/5530 Data Science/A1-AmyNgo/Student-Performance-Visuals/results/visual-output'
```

```python
avg_scores = clean_student_data.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

sns.set_theme(style="whitegrid")

bar_width = 0.25
positions = np.arange(len(avg_scores.index))

plt.figure(figsize=(14, 8))

plt.bar(positions - bar_width, avg_scores['math score'], 
        width=bar_width, color='orange', label='Math Score')
plt.bar(positions, avg_scores['reading score'], 
        width=bar_width, color='purple', label='Reading Score')
plt.bar(positions + bar_width, avg_scores['writing score'], 
        width=bar_width, color='lightgreen', label='Writing Score')

plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')

plt.xticks(ticks=positions, labels=avg_scores.index, rotation=45)
plt.legend(title='Score Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
#plt.savefig(os.path.join(save_folder, 'avg_scores_by_race.png'))

plt.show()

```

```python
sns.set_theme(style="dark")

fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharey=True)

sns.histplot(clean_student_data['math score'], color='blue', kde=True, bins=20, ax=axes[0])
axes[0].set_title('Distribution of Math Scores')
axes[0].set_xlabel('Math Score')
axes[0].set_ylabel('Frequency')

sns.histplot(clean_student_data['reading score'], color='green', kde=True, bins=20, ax=axes[1])
axes[1].set_title('Distribution of Reading Scores')
axes[1].set_xlabel('Reading Score')
axes[1].set_ylabel('Frequency')

sns.histplot(clean_student_data['writing score'], color='red', kde=True, bins=20, ax=axes[2])
axes[2].set_title('Distribution of Writing Scores')
axes[2].set_xlabel('Writing Score')
axes[2].set_ylabel('Frequency')

for ax in axes:
    ax.grid(True)

plt.tight_layout()
#plt.savefig(os.path.join(save_folder, 'score_distributions.png'))

plt.show()
```

```python
sns.set_theme(style="dark")

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

sns.boxplot(x='test preparation course', 
            y='math score', 
            data=clean_student_data, 
            palette='husl', 
            ax=axes[0])
axes[0].set_title('Math Scores by Test Preparation Course')
axes[0].set_xlabel('Test Preparation')
axes[0].set_ylabel('Math Score')

sns.boxplot(x='test preparation course', 
            y='reading score', 
            data=clean_student_data, 
            palette='husl', 
            ax=axes[1])
axes[1].set_title('Reading Scores by Test Preparation Course')
axes[1].set_xlabel('Test Preparation')
axes[1].set_ylabel('Reading Score')

sns.boxplot(x='test preparation course', 
            y='writing score', 
            data=clean_student_data, 
            palette='husl', 
            ax=axes[2])
axes[2].set_title('Writing Scores by Test Preparation Course')
axes[2].set_xlabel('Test Preparation')
axes[2].set_ylabel('Writing Score')

for ax in axes:
    ax.grid(True)

plt.tight_layout()
#plt.savefig(os.path.join(save_folder, 'scores_by_test_prep.png'))

plt.show()
```

```python
average_scores = clean_student_data.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

sns.set_theme(style="whitegrid")

bar_width = 0.25
positions = np.arange(len(average_scores.index))

plt.figure(figsize=(10, 6))

plt.bar(positions - bar_width, average_scores['math score'], 
        width=bar_width, color='orange', label='Math Score')
plt.bar(positions, average_scores['reading score'], 
        width=bar_width, color='purple', label='Reading Score')
plt.bar(positions + bar_width, average_scores['writing score'], 
        width=bar_width, color='lightgreen', label='Writing Score')

plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(ticks=positions, labels=average_scores.index, rotation=45)

plt.legend(title='Score Type',loc='upper left',bbox_to_anchor=(1, 1))
plt.grid(True)
#plt.savefig(os.path.join(save_folder, 'avg_scores_by_parentEd.png'))

plt.show()
```

```python
education_counts = clean_student_data.groupby(['race/ethnicity', 'parental level of education']).size().unstack()

sns.set_theme(style="dark")

education_counts.plot(kind='bar', 
                      figsize=(14, 8), 
                      color=['skyblue', 'orange', 'green', 'red', 'purple', 'brown'], 
                      edgecolor='black')

plt.title('Parental Education Levels by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count of Students')

plt.xticks(rotation=45)
plt.legend(title='Parental Education', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
#plt.savefig(os.path.join(save_folder, 'parentEd_by_race.png'))

plt.show()
```
